### 12) WAP to count occurrences of an element in a list.

```python
l1=[1,2,3,4,5,6,1]
d={}
for i in l1:
    if i not in d:
        d[i]=1
    else:
        d[i]+=1
print(d)
```

### 14) WAP to extract elements with frequency greater than K.

```python
k=int(input("enter k: "))
l1=[1,2,3,4,5,6,1]
d={}
for i in l1:
    if i not in d:
        d[i]=1
    else:
        d[i]+=1
for j,v in d.items():
    if v==k:
        print(j)
```
