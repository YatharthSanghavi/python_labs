### 7. WAP to print sum of series 1 – 2 + 3 – 4 + 5 – 6 + 7 ... n.

```python
n1=int(input("enter n: "))
sum=0
for i in range(1,n1+1):
    if i%2==0:
        sum-=i
    else:
        sum+=i
print(sum)
```

### 12. WAP to print sum of digits of given number.

```python
n1=int(input("enter n: "))
sum1=0
while n1>0:
    digit=n1%10
    sum1+=digit
    n1/=10
print(int(sum1))
```

### 13. WAP to check whether the given number is palindrome or not.

```python
n1=input("enter n: ")
n2=int(n1[::-1])
n1=int(n1)
if n1==n2:
    print("palindrome")
```

### 15. Armstrong Number

```python
n1=int(input("enter n: "))
l=len(str(n1))
sum=0
t=n1
while n1>0:
    div=n1%10
    sum+= div**l
    n1//=10
if sum == t:
    print("it is Armstrong num")
else:
    print("not armstrong num")
```
